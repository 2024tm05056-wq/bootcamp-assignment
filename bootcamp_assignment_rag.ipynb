### 1. Domain Identification
The identified domain is **Climate Science, Environmental Policy, and Sustainability**. ### 2. Problem Description
The primary challenge within this domain, particularly concerning comprehensive reports like the IPCC SYR, is the sheer volume, density, and diverse representation of information. Climate scientists and policymakers frequently face an information overload, making it exceedingly difficult and time-consuming to extract precise, evidence-based answers to complex, multi-faceted queries. Current manual methods for navigating these extensive documents, which include thousands of pages of text, intricate tabular data, and highly informative figures (e.g., graphs, diagrams, maps), are inefficient and prone to human error or oversight. Users often struggle to:

*   **Synthesize information across modalities**: Key insights are often distributed across textual explanations, numerical data in tables, and visual trends in figures. For instance, understanding the projected impact of a specific warming level on regional water availability might require reading descriptive text, cross-referencing a table of climate model projections, and analyzing a geographical map.
*   **Locate specific quantitative data**: Extracting exact numerical values or trends from tables embedded within the report is cumbersome.
*   **Interpret visual information**: Understanding the implications of complex figures, beyond their captions, often requires detailed contextual reading.
*   **Verify scientific claims**: Policymakers require swift validation of statements by tracing them directly back to the report's evidence, which is challenging when the information source spans multiple pages and content types.

This fragmentation of critical information across different formats hinders rapid evidence-based decision-making and limits the full utilization of the rich data contained within these pivotal climate assessment reports.

The uniqueness of this problem stems from the specific characteristics of climate change reports and the requirements of their users:

*   **High-Stakes Multimodality**: Unlike many other domains where text is the predominant information carrier, climate science demands the integration of textual narratives with precise quantitative data (tables) and complex visual representations (figures) to convey scientific consensus, uncertainties, and projections. The misinterpretation or omission of information from any modality can have severe global consequences.
*   **Nuance and Specificity**: Queries in climate science often require highly specific, nuanced answers that factor in geographic, temporal, and socio-economic contexts. A simple keyword search is insufficient; semantic understanding of the query and the document's content is crucial.

*   **Evolving Understanding of Climate Science**: While the report itself is static, the scientific understanding it encapsulates is constantly evolving. A system that can reliably extract and synthesize information from such a foundational report can serve as a robust baseline for integrating future updates or comparing with new research.

### 4. Why Retrieval-Augmented Generation (RAG) is the Best Approach
Retrieval-Augmented Generation (RAG) offers a superior solution compared to standalone Large Language Models (LLMs) or traditional information retrieval systems for this problem due to several key advantages:

*   **Mitigation of Hallucination**: Pure LLMs are prone to generating plausible but incorrect information. RAG directly addresses this by grounding the LLM's response in factual evidence retrieved from the authoritative IPCC report, significantly improving accuracy and trustworthiness.
*   **Multimodal Information Integration**: RAG can be architected to retrieve information not just from text, but also from structured tables and interpret visual cues/metadata from figures. For instance, a query about "sea level rise projections for small island states" could retrieve relevant text passages, a table with specific projection values, and a graph visualizing the trend, all contributing to a comprehensive answer.
*   **Attribution and Verifiability**: A core strength of RAG is its ability to provide explicit citations or references (e.g., page numbers, table IDs, figure numbers) for the retrieved information. This is critical for scientific and policy verification, allowing users to trace the generated answers back to the original report, fostering transparency and trust.
*   **Handling Complex and Long Documents**: RAG excels at navigating extensive documents by retrieving only the most relevant chunks of information, rather than processing the entire report. This makes it efficient for answering specific questions within a lengthy and dense publication.
*   **Semantic Understanding**: RAG leverages the advanced natural language understanding capabilities of LLMs to interpret complex queries and semantically match them with relevant content, even if exact keywords are not present.
*   **Scalability**: The RAG framework is inherently scalable. Once established for a single report, it can be extended to index and retrieve information from multiple IPCC reports or other climate-related documents, building a vast, verifiable knowledge base.

## 9. Build the Retrieval-Augmented Generation (RAG) Pipeline

This section integrates the retrieval component (FAISS index and `retrieve_similar_chunks` function) with a Large Language Model (LLM) to answer user queries using only the retrieved context. The goal is to provide grounded answers and include source references to prevent hallucination.

````python
import openai
import os
from google.colab import userdata # For securely accessing API keys
from google.colab.userdata import SecretNotFoundError # Import the specific exception

# Set your OpenAI API key from Colab secrets
# You need to add OPENAI_API_KEY to Colab secrets under the '🔑' icon on the left panel.
try:
    openai_api_key = userdata.get('OPENAI_API_KEY')
    openai.api_key = openai_api_key
except SecretNotFoundError:
    print("Error: OPENAI_API_KEY not found in Colab secrets. Please add it via the '🔑' icon on the left panel.")
    openai_api_key = None
except Exception as e:
    print(f"An unexpected error occurred while accessing Colab secrets: {e}")
    openai_api_key = None

def answer_query_with_rag(query, k=5, llm_model="gpt-3.5-turbo"):
    """
    Orchestrates the RAG pipeline to answer a query using retrieved chunks and an LLM.

    Args:
        query (str): The user's query.
        k (int): Number of top-k chunks to retrieve.
        llm_model (str): The name of the OpenAI LLM model to use.

    Returns:
        str: The LLM-generated answer with source references.
    """
    if not openai_api_key:
        return "Cannot generate answer: OpenAI API key is not set or accessible."

    print(f"Retrieving top {k} chunks for the query: '{query}'")
    retrieved_chunks = retrieve_similar_chunks(query, k=k)

    # 1. Combine retrieved chunks into context
    context = ""
    sources = set() # To store unique source references

    for i, chunk in enumerate(retrieved_chunks):
        context += f"---\nContext Chunk {i+1} (Type: {chunk['chunk_type']}, Page: {chunk['page_num']}, ID: {chunk['chunk_id']}):\n{chunk['content']}\n"
        sources.add(f"[Page {chunk['page_num']}, {chunk['chunk_type'].capitalize()} ID: {chunk['chunk_id']}]")

    # 2. Create a custom prompt template
    prompt = f"""You are an intelligent assistant designed to answer questions based ONLY on the provided context. If the answer cannot be found in the context, clearly state 'I cannot answer this question based on the provided information.' Do not make up information.

Provide a concise and accurate answer. ALWAYS cite the source(s) from the context that directly support your answer. Cite using the format [Page X, Type Y ID: Z].

Context:
{context}

Question: {query}

Answer:"""

    print("\nSending query and context to LLM...")
    try:
        # 3. Send to LLM
        response = openai.ChatCompletion.create(
            model=llm_model,
            messages=[
                {"role": "system", "content": "You are a helpful assistant."},
                {"role": "user", "content": prompt}
            ],
            temperature=0.0 # Keep temperature low for factual, grounded answers
        )

        llm_answer = response.choices[0].message.content.strip()

        # 4. Generate final answer and include source references
        # The prompt encourages the LLM to cite, but we can also add a general list at the end.
        final_answer = llm_answer
        if not any(source in llm_answer for source in sources):
            final_answer += f"\n\nSources used: {'; '.join(sorted(list(sources)))}"

        return final_answer

    except Exception as e:
        return f"Error communicating with LLM: {e}"

print("RAG pipeline function 'answer_query_with_rag' is now available.")
print("Example usage: answer_query_with_rag(\"What are the main risks associated with global warming of 1.5°C or higher?\")")
````

````python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
import json
import os

# 1. Combine all chunks (text, image captions, tables)
all_chunks = []

# Add text chunks
for chunk in text_chunks:
    all_chunks.append({
        'content': chunk['content'],
        'page_num': chunk['page_num'],
        'chunk_id': chunk['chunk_id'],
        'chunk_type': 'text'
    })

# Add image caption chunks
for chunk in image_caption_chunks:
    all_chunks.append({
        'content': chunk['content'],
        'page_num': chunk['page_num'],
        'chunk_id': chunk['chunk_id'],
        'chunk_type': 'image_caption'
    })

# Add table chunks
for chunk in table_chunks:
    all_chunks.append({
        'content': chunk['content'],
        'page_num': chunk['page_num'],
        'chunk_id': chunk['chunk_id'],
        'chunk_type': 'table'
    })

print(f"Total number of combined chunks: {len(all_chunks)}")

# 2. Load embedding model
# 'all-MiniLM-L6-v2' is a good balance of performance and speed.
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')

# 3. Generate embeddings for all chunks
chunk_contents = [chunk['content'] for chunk in all_chunks]
chunk_embeddings = embedding_model.encode(chunk_contents, show_progress_bar=True)

print(f"Generated embeddings for {len(chunk_embeddings)} chunks. Embedding dimension: {chunk_embeddings.shape[1]}")

# 4. Create FAISS index
dimension = chunk_embeddings.shape[1]
faiss_index = faiss.IndexFlatL2(dimension) # Using L2 distance for similarity search

# Add embeddings to the index
faiss_index.add(np.array(chunk_embeddings).astype('float32'))

print(f"FAISS index created with {faiss_index.ntotal} embeddings.")

# 5. Store metadata and FAISS index for persistence
metadata_path = os.path.join(project_dir, 'vector_db', 'all_chunks_metadata.json')
with open(metadata_path, 'w') as f:
    json.dump(all_chunks, f, indent=4)

faiss_index_path = os.path.join(project_dir, 'vector_db', 'faiss_index.bin')
faiss.write_index(faiss_index, faiss_index_path)

print(f"FAISS index saved to: {faiss_index_path}")
print(f"Chunk metadata saved to: {metadata_path}")

# 6. Implement Similarity Search Function
def retrieve_similar_chunks(query, k=5):
    """
    Retrieves the top k most similar chunks from the FAISS index for a given query.

    Args:
        query (str): The query string.
        k (int): The number of top similar chunks to retrieve.

    Returns:
        list: A list of dictionaries, each containing information about a retrieved chunk.
    """
    query_embedding = embedding_model.encode([query]).astype('float32')
    distances, indices = faiss_index.search(query_embedding, k)

    retrieved_info = []
    for i, idx in enumerate(indices[0]):
        # Ensure the index is valid (FAISS might return -1 if k > ntotal or other issues)
        if 0 <= idx < len(all_chunks):
            chunk = all_chunks[idx]
            retrieved_info.append({
                'rank': i + 1,
                'distance': distances[0][i],
                'chunk_id': chunk['chunk_id'],
                'chunk_type': chunk['chunk_type'],
                'page_num': chunk['page_num'],
                'content': chunk['content']
            })
        else:
            print(f"Warning: Invalid index {idx} returned by FAISS.")
    return retrieved_info

print("\nRetrieval function 'retrieve_similar_chunks' is ready.")
print("You can now test it with a query, for example:")
print("query_results = retrieve_similar_chunks(\"What are the global warming levels and their impacts?\")")
print("for res in query_results: print(res)\n")
````

## 6. Image Captioning

This step involves using a Vision Language Model (VLM) to generate descriptive captions for each extracted image. These captions will be treated as additional text chunks, allowing them to be embedded and retrieved alongside the text content from the PDF.

````python
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image
import os

# Load pre-trained BLIP model and processor
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model_blip = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

def generate_image_captions(images_info):
    image_caption_chunks = []
    for i, img_info in enumerate(images_info):
        image_path = img_info['image_path']
        page_num = img_info['page_num']

        if not os.path.exists(image_path):
            print(f"Warning: Image not found at {image_path}. Skipping.")
            continue

        raw_image = Image.open(image_path).convert('RGB')

        # Conditional image captioning (you can also use unconditional, but conditional is often better)
        text = "a photography of"
        inputs = processor(raw_image, text, return_tensors="pt")

        out = model_blip.generate(**inputs)
        caption = processor.decode(out[0], skip_special_tokens=True)

        image_caption_chunks.append({
            'content': f"Image from page {page_num}: {caption}",
            'page_num': page_num,
            'chunk_id': f'image_{i+1}_page_{page_num}_caption'
        })
        print(f"Generated caption for {img_info['filename']} (Page {page_num}): {caption}")

    return image_caption_chunks

# Assuming extracted_images_info is available from previous steps
image_caption_chunks = generate_image_captions(extracted_images_info)

print(f"\nCreated {len(image_caption_chunks)} image caption chunks.")
print("First caption chunk example:")
if image_caption_chunks:
    print(image_caption_chunks[0]['content'])
    print(f"Page number for first caption chunk: {image_caption_chunks[0]['page_num']}")
````

````python
import uvicorn
from fastapi import FastAPI, HTTPException, status, UploadFile, File
from pydantic import BaseModel
from typing import List, Dict, Any, Optional
import os
import json
import faiss
import numpy as np
from sentence_transformers import SentenceTransformer
import openai
from google.colab import userdata # Colab-specific secret handling
from google.colab.userdata import SecretNotFoundError
import fitz # PyMuPDF
import camelot
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image
from contextlib import asynccontextmanager # Import for lifespan management

# --- Configuration and Global Variables (Simulating the notebook state) ---
# In a real production app, these would be managed more robustly (e.g., config files, dependency injection)
# and possibly loaded on startup rather than being defined directly here.
# For now, we are using the project_dir already setup in the notebook.
PROJECT_DIR = '/content/multimodal_rag_project'

# Initialize these to None, they will be loaded/populated at startup or during ingest
embedding_model = None
faiss_index = None
all_chunks: List[Dict[str, Any]] = [] # Stores chunk metadata

# Blip model for image captioning
processor = None
model_blip = None

# OpenAI API key
OPENAI_API_KEY = None


# --- Pydantic Models ---
class HealthCheckResponse(BaseModel):
    status: str = "ok"
    message: str = "FastAPI application is healthy"

class QueryRequest(BaseModel):
    query_string: str
    k: int = 5

class QueryResult(BaseModel):
    answer: str
    sources: List[str]

class IngestResponse(BaseModel):
    message: str
    total_chunks_processed: int
    pdf_filename: str

# --- RAG Pipeline Helper Functions (Adapted from notebook cells) ---

def extract_pdf_content(pdf_file_path: str, output_project_dir: str):
    """Extracts text and images from a PDF using PyMuPDF."""
    doc = fitz.open(pdf_file_path)
    text_content_per_page = []
    images_info = []

    text_output_dir = os.path.join(output_project_dir, 'data', 'extracted_text')
    images_output_dir = os.path.join(output_project_dir, 'data', 'extracted_images')
    os.makedirs(text_output_dir, exist_ok=True)
    os.makedirs(images_output_dir, exist_ok=True)

    text_output_filename = os.path.basename(pdf_file_path).replace('.pdf', '_full_text.txt')
    text_output_path = os.path.join(text_output_dir, text_output_filename)

    with open(text_output_path, 'w', encoding='utf-8') as text_file:
        for page_num in range(doc.page_count):
            page = doc[page_num]
            page_text = page.get_text("text")
            text_content_per_page.append(f"--- Page {page_num + 1} ---\n{page_text}")
            text_file.write(f"--- Page {page_num + 1} ---\n{page_text}")

            image_list = page.get_images(full=True)
            for img_index, img in enumerate(image_list):
                xref = img[0]
                base_image = doc.extract_image(xref)
                if not base_image:
                    continue # Skip if image extraction fails
                image_bytes = base_image["image"]
                image_ext = base_image["ext"]
                image_filename = f"page{page_num + 1}_img{img_index}.{image_ext}"
                image_path = os.path.join(images_output_dir, image_filename)
                with open(image_path, "wb") as img_file:
                    img_file.write(image_bytes)
                images_info.append({
                    'page_num': page_num + 1,
                    'image_path': image_path,
                    'filename': image_filename
                })
    doc.close()
    return text_content_per_page, images_info

def extract_tables_from_pdf(pdf_file_path: str, output_project_dir: str):
    """Extracts tables from a PDF using Camelot."""
    extracted_tables = []
    tables_output_dir = os.path.join(output_project_dir, 'data', 'extracted_tables')
    os.makedirs(tables_output_dir, exist_ok=True)

    try:
        tables = camelot.read_pdf(pdf_file_path, pages='all', flavor='stream', split_text=True)
        for i, table in enumerate(tables):
            page_num = table.page
            df = table.df
            table_text_representation = f"Table from page {page_num}:\n" + df.to_string(index=False)
            table_filename_csv = f"page{page_num}_table{i+1}.csv"
            table_path_csv = os.path.join(tables_output_dir, table_filename_csv)
            df.to_csv(table_path_csv, index=False)
            extracted_tables.append({
                'content': table_text_representation,
                'page_num': page_num,
                'chunk_id': f'table_{i+1}_page_{page_num}',
                'filename': table_filename_csv,
                'path': table_path_csv,
            })
    except Exception as e:
        print(f"Error extracting tables: {e}") # Log the error
    return extracted_tables

def chunk_text(text_list: List[str], chunk_size: int = 500, chunk_overlap: int = 100):
    """Chunks raw text content into smaller pieces."""
    chunks = []
    for page_num_idx, page_content in enumerate(text_list):
        page_num = page_num_idx + 1 # Adjust for 1-based page numbering
        words = page_content.split()
        if len(words) <= chunk_size:
            chunks.append({
                'content': page_content,
                'page_num': page_num,
                'chunk_id': f'page_{page_num}_chunk_0'
            })
        else:
            for i in range(0, len(words), chunk_size - chunk_overlap):
                chunk_words = words[i:i + chunk_size]
                chunk_text = ' '.join(chunk_words)
                chunks.append({
                    'content': chunk_text,
                    'page_num': page_num,
                    'chunk_id': f'page_{page_num}_chunk_{i // (chunk_size - chunk_overlap)}'
                })
    return chunks

def generate_image_captions(images_info: List[Dict[str, Any]]):
    """Generates captions for extracted images using BLIP model."""
    image_caption_chunks = []
    if not processor or not model_blip:
        print("Warning: BLIP model not loaded. Ensure startup_event ran successfully. Skipping image captioning.")
        return []

    for i, img_info in enumerate(images_info):
        image_path = img_info['image_path']
        page_num = img_info['page_num']

        if not os.path.exists(image_path):
            print(f"Warning: Image not found at {image_path}. Skipping.")
            continue

        try:
            raw_image = Image.open(image_path).convert('RGB')
            text = "a photography of"
            inputs = processor(raw_image, text, return_tensors="pt")
            out = model_blip.generate(**inputs)
            caption = processor.decode(out[0], skip_special_tokens=True)
            image_caption_chunks.append({
                'content': f"Image from page {page_num}: {caption}",
                'page_num': page_num,
                'chunk_id': f'image_{i+1}_page_{page_num}_caption'
            })
        except Exception as e:
            print(f"Error generating caption for {img_info['filename']}: {e}") # Log the error
    return image_caption_chunks

def build_and_save_faiss_index(text_chunks, image_caption_chunks, table_chunks, output_project_dir: str):
    """Combines all chunks, generates embeddings, and builds/saves the FAISS index."""
    global all_chunks, faiss_index, embedding_model

    all_chunks = []
    for chunk in text_chunks:
        all_chunks.append({**chunk, 'chunk_type': 'text'})
    for chunk in image_caption_chunks:
        all_chunks.append({**chunk, 'chunk_type': 'image_caption'})
    for chunk in table_chunks:
        all_chunks.append({**chunk, 'chunk_type': 'table'})

    if not embedding_model:
        raise RuntimeError("Embedding model not loaded. Ensure startup_event ran successfully.")

    if not all_chunks:
        print("Warning: No chunks to embed. FAISS index will be empty.")
        faiss_index = None # Ensure index is cleared if no chunks
        return 0

    chunk_contents = [chunk['content'] for chunk in all_chunks]
    chunk_embeddings = embedding_model.encode(chunk_contents, show_progress_bar=False) # No progress bar in API

    dimension = chunk_embeddings.shape[1]
    faiss_index = faiss.IndexFlatL2(dimension)
    faiss_index.add(np.array(chunk_embeddings).astype('float32'))

    # Ensure vector_db directory exists
    vector_db_dir = os.path.join(output_project_dir, 'vector_db')
    os.makedirs(vector_db_dir, exist_ok=True)

    metadata_path = os.path.join(vector_db_dir, 'all_chunks_metadata.json')
    faiss_index_path = os.path.join(vector_db_dir, 'faiss_index.bin')

    with open(metadata_path, 'w') as f:
        json.dump(all_chunks, f, indent=4)
    faiss.write_index(faiss_index, faiss_index_path)

    return len(all_chunks)

def retrieve_similar_chunks(query: str, k: int = 5):
    """Retrieves the top k most similar chunks from the FAISS index for a given query."""
    if not faiss_index or not all_chunks or not embedding_model:
        raise RuntimeError("FAISS index or chunks not loaded. Please ingest a document first.")

    query_embedding = embedding_model.encode([query]).astype('float32')
    distances, indices = faiss_index.search(query_embedding, k)

    retrieved_info = []
    for i, idx in enumerate(indices[0]):
        # Ensure the index is valid (FAISS might return -1 if k > ntotal or other issues)
        if 0 <= idx < len(all_chunks):
            chunk = all_chunks[idx]
            retrieved_info.append({
                'rank': i + 1,
                'distance': distances[0][i],
                'chunk_id': chunk['chunk_id'],
                'chunk_type': chunk['chunk_type'],
                'page_num': chunk['page_num'],
                'content': chunk['content']
            })
        else:
            print(f"Warning: Invalid index {idx} returned by FAISS.") # Log this warning
    return retrieved_info

def answer_query_with_rag_fastapi(query: str, k: int = 5, llm_model: str = "gpt-3.5-turbo"):
    """Orchestrates the RAG pipeline to answer a query using retrieved chunks and an LLM."""
    global OPENAI_API_KEY
    if not OPENAI_API_KEY:
        return "Cannot generate answer: OpenAI API key is not set or accessible.", []

    retrieved_chunks = retrieve_similar_chunks(query, k=k)

    context = ""
    sources = set() # To store unique source references

    for i, chunk in enumerate(retrieved_chunks):
        context += f"---\nContext Chunk {i+1} (Type: {chunk['chunk_type']}, Page: {chunk['page_num']}, ID: {chunk['chunk_id']}):\n{chunk['content']}\n"
        sources.add(f"[Page {chunk['page_num']}, {chunk['chunk_type'].capitalize()} ID: {chunk['chunk_id']}]")

    prompt = f"""You are an intelligent assistant designed to answer questions based ONLY on the provided context. If the answer cannot be found in the context, clearly state 'I cannot answer this question based on the provided information.' Do not make up information.

Provide a concise and accurate answer. ALWAYS cite the source(s) from the context that directly support your answer. Cite using the format [Page X, Type Y ID: Z].

Context:
{context}

Question: {query}

Answer:"""

    try:
        # Ensure openai.api_key is set before making the call
        if openai.api_key is None:
            raise RuntimeError("OpenAI API key not set. Cannot communicate with LLM.")

        response = openai.ChatCompletion.create(
            model=llm_model,
            messages=[
                {"role": "system", "content": "You are a helpful assistant."},
                {"role": "user", "content": prompt}
            ],
            temperature=0.0 # Keep temperature low for factual, grounded answers
        )
        llm_answer = response.choices[0].message.content.strip()

        final_sources_list = sorted(list(sources))
        # The prompt encourages the LLM to cite, but we can also add a general list at the end if the LLM doesn't.
        if not any(source.replace('[', '').replace(']', '') in llm_answer for source in final_sources_list):
             llm_answer += f"\n\nSources used: {'; '.join(final_sources_list)}"

        return llm_answer, final_sources_list

    except Exception as e:
        print(f"Error communicating with LLM: {e}")
        return f"Error communicating with LLM: {e}", []


# --- FastAPI Application Setup ---

# Use lifespan context manager for startup/shutdown events
@asynccontextmanager
async def lifespan(app: FastAPI):
    global embedding_model, processor, model_blip, OPENAI_API_KEY
    print("Application startup: Initializing RAG components...")

    # Create project directory structure if it doesn't exist
    subdirs = [
        'data/raw_pdfs', 'data/extracted_text', 'data/extracted_images',
        'vector_db', 'models', 'output'
    ]
    for subdir in subdirs:
        os.makedirs(os.path.join(PROJECT_DIR, subdir), exist_ok=True)

    # Load embedding model
    try:
        embedding_model = SentenceTransformer('all-MiniLM-L6-v2')
        print("Sentence Transformer model loaded.")
    except Exception as e:
        print(f"Error loading Sentence Transformer model: {e}")
        embedding_model = None

    # Load BLIP model for image captioning
    try:
        processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
        model_blip = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")
        print("BLIP image captioning model loaded.")
    except Exception as e:
        print(f"Error loading BLIP model: {e}")
        processor = None
        model_blip = None

    # Get OpenAI API Key
    try:
        OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')
        openai.api_key = OPENAI_API_KEY
        print("OpenAI API key loaded from Colab secrets.")
    except SecretNotFoundError:
        print("Warning: OPENAI_API_KEY not found in Colab secrets. RAG queries will fail.")
        OPENAI_API_KEY = None
    except Exception as e:
        print(f"An unexpected error occurred while accessing Colab secrets: {e}")
        OPENAI_API_KEY = None

    print("RAG components initialization complete.")
    yield # Application starts here
    print("Application shutdown: Cleaning up resources (if any).")

app = FastAPI(
    title="Multimodal RAG API",
    description="API for a Retrieval-Augmented Generation system using text, tables, and images from PDF documents.",
    version="1.0.0",
    docs_url="/docs",
    redoc_url=None, # Disable Redoc if only Swagger UI is needed
    lifespan=lifespan # Assign the lifespan context manager
)


# --- FastAPI Endpoints ---

@app.get("/health", response_model=HealthCheckResponse, summary="Health Check")
async def health_check():
    """Checks if the API is running and healthy."""
    return HealthCheckResponse()

@app.post("/ingest", response_model=IngestResponse, status_code=status.HTTP_200_OK, summary="Ingest Document for RAG")
async def ingest_document(pdf_file: UploadFile = File(..., description="PDF file to ingest")):
    """Ingests a PDF document, extracts its content (text, tables, images), generates captions, and rebuilds the FAISS index.
    Note: This will replace any previously ingested document's index."""
    if not embedding_model or not processor or not model_blip:
        raise HTTPException(
            status_code=status.HTTP_500_INTERNAL_SERVER_ERROR,
            detail="RAG components (embedding/BLIP models) not loaded. Please ensure application started correctly."
        )

    # Save the uploaded PDF temporarily
    raw_pdfs_dir = os.path.join(PROJECT_DIR, 'data', 'raw_pdfs')
    os.makedirs(raw_pdfs_dir, exist_ok=True)
    temp_pdf_path = os.path.join(raw_pdfs_dir, pdf_file.filename)

    try:
        with open(temp_pdf_path, "wb") as buffer:
            buffer.write(await pdf_file.read())
        print(f"Uploaded PDF saved to: {temp_pdf_path}")

        # Process the PDF
        print("Starting PDF content extraction...")
        full_text_list, extracted_images_info = extract_pdf_content(temp_pdf_path, PROJECT_DIR)
        print(f"Extracted {len(full_text_list)} pages of text and {len(extracted_images_info)} images.")

        text_chunks = chunk_text(full_text_list)
        print(f"Created {len(text_chunks)} text chunks.")

        table_chunks = extract_tables_from_pdf(temp_pdf_path, PROJECT_DIR)
        print(f"Extracted {len(table_chunks)} table chunks.")

        image_caption_chunks = generate_image_captions(extracted_images_info)
        print(f"Generated {len(image_caption_chunks)} image caption chunks.")

        total_chunks = build_and_save_faiss_index(text_chunks, image_caption_chunks, table_chunks, PROJECT_DIR)
        print(f"FAISS index rebuilt with {total_chunks} chunks.")

        return IngestResponse(
            message=f"Successfully processed {pdf_file.filename} and rebuilt FAISS index.",
            total_chunks_processed=total_chunks,
            pdf_filename=pdf_file.filename
        )
    except Exception as e:
        # Log the detailed exception for debugging
        import traceback
        traceback.print_exc()
        raise HTTPException(status_code=status.HTTP_500_INTERNAL_SERVER_ERROR, detail=f"Failed to ingest document: {e}")
    finally:
        # Clean up temporary file
        if os.path.exists(temp_pdf_path):
            os.remove(temp_pdf_path)
            print(f"Cleaned up temporary PDF file: {temp_pdf_path}")

@app.post("/query", response_model=QueryResult, status_code=status.HTTP_200_OK, summary="Query RAG System")
async def query_rag(request: QueryRequest):
    """Queries the RAG system to get an answer based on the ingested document and cited sources."""
    if not faiss_index or not all_chunks or not embedding_model:
        raise HTTPException(
            status_code=status.HTTP_503_SERVICE_UNAVAILABLE,
            detail="RAG index not initialized. Please ingest a document first via the /ingest endpoint."
        )
    if not OPENAI_API_KEY:
        raise HTTPException(
            status_code=status.HTTP_503_SERVICE_UNAVAILABLE,
            detail="OpenAI API key is not configured. Please set it in Colab secrets."
        )

    try:
        answer, sources_list = answer_query_with_rag_fastapi(request.query_string, request.k)
        if answer.startswith("Error communicating with LLM"):
             raise HTTPException(status_code=status.HTTP_500_INTERNAL_SERVER_ERROR, detail=answer)
        return QueryResult(answer=answer, sources=sources_list)
    except RuntimeError as e:
        raise HTTPException(status_code=status.HTTP_500_INTERNAL_SERVER_ERROR, detail=f"RAG system error: {e}")
    except Exception as e:
        # Log the detailed exception for debugging
        import traceback
        traceback.print_exc()
        raise HTTPException(status_code=status.HTTP_500_INTERNAL_SERVER_ERROR, detail=f"An unexpected error occurred during query: {e}")

# To run the FastAPI application in Colab
# You can uncomment and run this part in a separate cell
# if __name__ == "__main__":
#     # For Colab, you might need to use a tool like ngrok to expose the local server
#     # or simply run it locally and access via the displayed IP.
#     # Example using ngrok (install ngrok first: !pip install pyngrok)
#     # from pyngrok import ngrok
#     # public_url = ngrok.connect(8000)
#     # print(f"Public URL: {public_url}")
#     uvicorn.run(app, host="0.0.0.0", port=8000)
````

This module will contain functions responsible for extracting content (text, tables, images) from PDFs and processing them into chunks or captions.

````python
%%writefile /content/multimodal_rag_project/src/ingestion/pdf_processor.py

import os
import fitz # PyMuPDF
import camelot
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image
from typing import List, Dict, Any

# Global BLIP models (will be loaded at API startup)
processor = None
model_blip = None

def set_blip_models(p: BlipProcessor, m: BlipForConditionalGeneration):
    global processor, model_blip
    processor = p
    model_blip = m

def extract_pdf_content(pdf_file_path: str, output_project_dir: str):
    """Extracts text and images from a PDF using PyMuPDF."""
    doc = fitz.open(pdf_file_path)
    text_content_per_page = []
    images_info = []

    text_output_dir = os.path.join(output_project_dir, 'data', 'extracted_text')
    images_output_dir = os.path.join(output_project_dir, 'data', 'extracted_images')
    os.makedirs(text_output_dir, exist_ok=True)
    os.makedirs(images_output_dir, exist_ok=True)

    text_output_filename = os.path.basename(pdf_file_path).replace('.pdf', '_full_text.txt')
    text_output_path = os.path.join(text_output_dir, text_output_filename)

    with open(text_output_path, 'w', encoding='utf-8') as text_file:
        for page_num in range(doc.page_count):
            page = doc[page_num]
            page_text = page.get_text("text")
            text_content_per_page.append(f"--- Page {page_num + 1} ---\n{page_text}")
            text_file.write(f"--- Page {page_num + 1} ---\n{page_text}")

            image_list = page.get_images(full=True)
            for img_index, img in enumerate(image_list):
                xref = img[0]
                base_image = doc.extract_image(xref)
                if not base_image:
                    continue # Skip if image extraction fails
                image_bytes = base_image["image"]
                image_ext = base_image["ext"]
                image_filename = f"page{page_num + 1}_img{img_index}.{image_ext}"
                image_path = os.path.join(images_output_dir, image_filename)
                with open(image_path, "wb") as img_file:
                    img_file.write(image_bytes)
                images_info.append({
                    'page_num': page_num + 1,
                    'image_path': image_path,
                    'filename': image_filename
                })
    doc.close()
    return text_content_per_page, images_info

def extract_tables_from_pdf(pdf_file_path: str, output_project_dir: str):
    """Extracts tables from a PDF using Camelot."""
    extracted_tables = []
    tables_output_dir = os.path.join(output_project_dir, 'data', 'extracted_tables')
    os.makedirs(tables_output_dir, exist_ok=True)

    try:
        tables = camelot.read_pdf(pdf_file_path, pages='all', flavor='stream', split_text=True)
        for i, table in enumerate(tables):
            page_num = table.page
            df = table.df
            table_text_representation = f"Table from page {page_num}:\n" + df.to_string(index=False)
            table_filename_csv = f"page{page_num}_table{i+1}.csv"
            table_path_csv = os.path.join(tables_output_dir, table_filename_csv)
            df.to_csv(table_path_csv, index=False)
            extracted_tables.append({
                'content': table_text_representation,
                'page_num': page_num,
                'chunk_id': f'table_{i+1}_page_{page_num}',
                'filename': table_filename_csv,
                'path': table_path_csv,
            })
    except Exception as e:
        print(f"Error extracting tables: {e}") # Log the error
    return extracted_tables

def chunk_text(text_list: List[str], chunk_size: int = 500, chunk_overlap: int = 100):
    """Chunks raw text content into smaller pieces."""
    chunks = []
    for page_num_idx, page_content in enumerate(text_list):
        page_num = page_num_idx + 1 # Adjust for 1-based page numbering
        words = page_content.split()
        if len(words) <= chunk_size:
            chunks.append({
                'content': page_content,
                'page_num': page_num,
                'chunk_id': f'page_{page_num}_chunk_0'
            })
        else:
            for i in range(0, len(words), chunk_size - chunk_overlap):
                chunk_words = words[i:i + chunk_size]
                chunk_text = ' '.join(chunk_words)
                chunks.append({
                    'content': chunk_text,
                    'page_num': page_num,
                    'chunk_id': f'page_{page_num}_chunk_{i // (chunk_size - chunk_overlap)}'
                })
    return chunks

def generate_image_captions(images_info: List[Dict[str, Any]]):
    """Generates captions for extracted images using BLIP model."""
    image_caption_chunks = []
    if not processor or not model_blip:
        print("Warning: BLIP model not loaded. Ensure it's set via set_blip_models(). Skipping image captioning.")
        return []

    for i, img_info in enumerate(images_info):
        image_path = img_info['image_path']
        page_num = img_info['page_num']

        if not os.path.exists(image_path):
            print(f"Warning: Image not found at {image_path}. Skipping.")
            continue

        try:
            raw_image = Image.open(image_path).convert('RGB')
            text = "a photography of"
            inputs = processor(raw_image, text, return_tensors="pt")
            out = model_blip.generate(**inputs)
            caption = processor.decode(out[0], skip_special_tokens=True)
            image_caption_chunks.append({
                'content': f"Image from page {page_num}: {caption}",
                'page_num': page_num,
                'chunk_id': f'image_{i+1}_page_{page_num}_caption'
            })
        except Exception as e:
            print(f"Error generating caption for {img_info['filename']}: {e}") # Log the error
    return image_caption_chunks
````

### `src/retrieval/vector_db.py`

This module will handle the creation, storage, and retrieval of vector embeddings using FAISS.

````python
%%writefile /content/multimodal_rag_project/src/retrieval/vector_db.py

import faiss
import numpy as np
import json
import os
from sentence_transformers import SentenceTransformer
from typing import List, Dict, Any

# Global variables for the FAISS index and chunk metadata
faiss_index = None
all_chunks: List[Dict[str, Any]] = []
embedding_model = None

def set_embedding_model(model: SentenceTransformer):
    global embedding_model
    embedding_model = model

def build_and_save_faiss_index(text_chunks, image_caption_chunks, table_chunks, output_project_dir: str):
    """Combines all chunks, generates embeddings, and builds/saves the FAISS index."""
    global all_chunks, faiss_index, embedding_model

    all_chunks = []
    for chunk in text_chunks:
        all_chunks.append({**chunk, 'chunk_type': 'text'})
    for chunk in image_caption_chunks:
        all_chunks.append({**chunk, 'chunk_type': 'image_caption'})
    for chunk in table_chunks:
        all_chunks.append({**chunk, 'chunk_type': 'table'})

    if not embedding_model:
        raise RuntimeError("Embedding model not loaded. Ensure it's set via set_embedding_model().")

    if not all_chunks:
        print("Warning: No chunks to embed. FAISS index will be empty.")
        faiss_index = None # Ensure index is cleared if no chunks
        return 0

    chunk_contents = [chunk['content'] for chunk in all_chunks]
    chunk_embeddings = embedding_model.encode(chunk_contents, show_progress_bar=False) # No progress bar in API

    dimension = chunk_embeddings.shape[1]
    faiss_index = faiss.IndexFlatL2(dimension)
    faiss_index.add(np.array(chunk_embeddings).astype('float32'))

    # Ensure vector_db directory exists
    vector_db_dir = os.path.join(output_project_dir, 'vector_db')
    os.makedirs(vector_db_dir, exist_ok=True)

    metadata_path = os.path.join(vector_db_dir, 'all_chunks_metadata.json')
    faiss_index_path = os.path.join(vector_db_dir, 'faiss_index.bin')

    with open(metadata_path, 'w') as f:
        json.dump(all_chunks, f, indent=4)
    faiss.write_index(faiss_index, faiss_index_path)

    return len(all_chunks)

def retrieve_similar_chunks(query: str, k: int = 5):
    """Retrieves the top k most similar chunks from the FAISS index for a given query."""
    if faiss_index is None or not all_chunks or embedding_model is None:
        raise RuntimeError("FAISS index or chunks not loaded. Please ingest a document first.")

    query_embedding = embedding_model.encode([query]).astype('float32')
    distances, indices = faiss_index.search(query_embedding, k)

    retrieved_info = []
    for i, idx in enumerate(indices[0]):
        # Ensure the index is valid (FAISS might return -1 if k > ntotal or other issues)
        if 0 <= idx < len(all_chunks):
            chunk = all_chunks[idx]
            retrieved_info.append({
                'rank': i + 1,
                'distance': distances[0][i],
                'chunk_id': chunk['chunk_id'],
                'chunk_type': chunk['chunk_type'],
                'page_num': chunk['page_num'],
                'content': chunk['content']
            })
        else:
            print(f"Warning: Invalid index {idx} returned by FAISS.") # Log this warning
    return retrieved_info
````

### `src/retrieval/rag_pipeline.py`

This module will contain the core RAG logic, integrating retrieval with the LLM.

````python
%%writefile /content/multimodal_rag_project/src/retrieval/rag_pipeline.py

import openai
from typing import List, Dict, Any
from src.retrieval.vector_db import retrieve_similar_chunks # Import the retrieval function

# Global variable for OpenAI API Key
OPENAI_API_KEY = None

def set_openai_api_key(key: str):
    global OPENAI_API_KEY
    OPENAI_API_KEY = key
    openai.api_key = key

def answer_query_with_rag_fastapi(query: str, k: int = 5, llm_model: str = "gpt-3.5-turbo"):
    """Orchestrates the RAG pipeline to answer a query using retrieved chunks and an LLM."""
    if not OPENAI_API_KEY:
        return "Cannot generate answer: OpenAI API key is not set or accessible.", []

    retrieved_chunks = retrieve_similar_chunks(query, k=k)

    context = ""
    sources = set() # To store unique source references

    for i, chunk in enumerate(retrieved_chunks):
        context += f"---\nContext Chunk {i+1} (Type: {chunk['chunk_type']}, Page: {chunk['page_num']}, ID: {chunk['chunk_id']}):\n{chunk['content']}\n"
        sources.add(f"[Page {chunk['page_num']}, {chunk['chunk_type'].capitalize()} ID: {chunk['chunk_id']}]")

    prompt = f"""You are an intelligent assistant designed to answer questions based ONLY on the provided context. If the answer cannot be found in the context, clearly state 'I cannot answer this question based on the provided information.' Do not make up information.

Provide a concise and accurate answer. ALWAYS cite the source(s) from the context that directly support your answer. Cite using the format [Page X, Type Y ID: Z].

Context:
{context}

Question: {query}

Answer:"""

    try:
        # Ensure openai.api_key is set before making the call
        if openai.api_key is None:
            raise RuntimeError("OpenAI API key not set. Cannot communicate with LLM.")

        response = openai.ChatCompletion.create(
            model=llm_model,
            messages=[
                {"role": "system", "content": "You are a helpful assistant."},
                {"role": "user", "content": prompt}
            ],
            temperature=0.0 # Keep temperature low for factual, grounded answers
        )
        llm_answer = response.choices[0].message.content.strip()

        final_sources_list = sorted(list(sources))
        # The prompt encourages the LLM to cite, but we can also add a general list at the end if the LLM doesn't.
        if not any(source.replace('[', '').replace(']', '') in llm_answer for source in final_sources_list):
             llm_answer += f"\n\nSources used: {'; '.join(final_sources_list)}"

        return llm_answer, final_sources_list

    except Exception as e:
        print(f"Error communicating with LLM: {e}")
        return f"Error communicating with LLM: {e}", []
````

````python
%%writefile /content/multimodal_rag_project/src/api/main.py

import uvicorn
from fastapi import FastAPI, HTTPException, status, UploadFile, File
from contextlib import asynccontextmanager
import os

from google.colab import userdata # Colab-specific secret handling
from google.colab.userdata import SecretNotFoundError

from sentence_transformers import SentenceTransformer
from transformers import BlipProcessor, BlipForConditionalGeneration

from src.models.api_models import HealthCheckResponse, QueryRequest, QueryResult, IngestResponse
from src.ingestion.pdf_processor import extract_pdf_content, extract_tables_from_pdf, chunk_text, generate_image_captions, set_blip_models
from src.retrieval.vector_db import build_and_save_faiss_index, set_embedding_model
from src.retrieval.rag_pipeline import answer_query_with_rag_fastapi, set_openai_api_key

# --- Configuration and Global Variables ---
PROJECT_DIR = '/content/multimodal_rag_project'

# --- Lifespan Context Manager ---
@asynccontextmanager
async def lifespan(app: FastAPI):
    # Initialize global components
    print("Application startup: Initializing RAG components...")

    # Create project directory structure if it doesn't exist
    subdirs = [
        'data/raw_pdfs', 'data/extracted_text', 'data/extracted_images',
        'vector_db', 'models', 'output'
    ]
    for subdir in subdirs:
        os.makedirs(os.path.join(PROJECT_DIR, subdir), exist_ok=True)

    # Load embedding model
    try:
        embedding_model_instance = SentenceTransformer('all-MiniLM-L6-v2')
        set_embedding_model(embedding_model_instance)
        print("Sentence Transformer model loaded.")
    except Exception as e:
        print(f"Error loading Sentence Transformer model: {e}")

    # Load BLIP model for image captioning
    try:
        processor_instance = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
        model_blip_instance = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")
        set_blip_models(processor_instance, model_blip_instance)
        print("BLIP image captioning model loaded.")
    except Exception as e:
        print(f"Error loading BLIP model: {e}")

    # Get OpenAI API Key
    try:
        openai_key = userdata.get('OPENAI_API_KEY')
        set_openai_api_key(openai_key)
        print("OpenAI API key loaded from Colab secrets.")
    except SecretNotFoundError:
        print("Warning: OPENAI_API_KEY not found in Colab secrets. RAG queries will fail.")
    except Exception as e:
        print(f"An unexpected error occurred while accessing Colab secrets: {e}")

    print("RAG components initialization complete.")
    yield # Application starts here
    print("Application shutdown: Cleaning up resources (if any).")

# --- FastAPI Application Instance ---
app = FastAPI(
    title="Multimodal RAG API",
    description="API for a Retrieval-Augmented Generation system using text, tables, and images from PDF documents.",
    version="1.0.0",
    docs_url="/docs",
    redoc_url=None, # Disable Redoc if only Swagger UI is needed
    lifespan=lifespan # Assign the lifespan context manager
)

# --- FastAPI Endpoints ---

@app.get("/health", response_model=HealthCheckResponse, summary="Health Check")
async def health_check():
    """Checks if the API is running and healthy."""
    return HealthCheckResponse()

@app.post("/ingest", response_model=IngestResponse, status_code=status.HTTP_200_OK, summary="Ingest Document for RAG")
async def ingest_document(pdf_file: UploadFile = File(..., description="PDF file to ingest")):
    """Ingests a PDF document, extracts its content (text, tables, images), generates captions, and rebuilds the FAISS index.
    Note: This will replace any previously ingested document's index."""
    # Check if models are loaded before proceeding
    from src.retrieval.vector_db import embedding_model
    from src.ingestion.pdf_processor import processor, model_blip

    if embedding_model is None or processor is None or model_blip is None:
        raise HTTPException(
            status_code=status.HTTP_500_INTERNAL_SERVER_ERROR,
            detail="RAG components (embedding/BLIP models) not loaded. Please ensure application started correctly."
        )

    # Save the uploaded PDF temporarily
    raw_pdfs_dir = os.path.join(PROJECT_DIR, 'data', 'raw_pdfs')
    os.makedirs(raw_pdfs_dir, exist_ok=True)
    temp_pdf_path = os.path.join(raw_pdfs_dir, pdf_file.filename)

    try:
        with open(temp_pdf_path, "wb") as buffer:
            buffer.write(await pdf_file.read())
        print(f"Uploaded PDF saved to: {temp_pdf_path}")

        # Process the PDF
        print("Starting PDF content extraction...")
        full_text_list, extracted_images_info = extract_pdf_content(temp_pdf_path, PROJECT_DIR)
        print(f"Extracted {len(full_text_list)} pages of text and {len(extracted_images_info)} images.")

        text_chunks = chunk_text(full_text_list)
        print(f"Created {len(text_chunks)} text chunks.")

        table_chunks = extract_tables_from_pdf(temp_pdf_path, PROJECT_DIR)
        print(f"Extracted {len(table_chunks)} table chunks.")

        image_caption_chunks = generate_image_captions(extracted_images_info)
        print(f"Generated {len(image_caption_chunks)} image caption chunks.")

        total_chunks = build_and_save_faiss_index(text_chunks, image_caption_chunks, table_chunks, PROJECT_DIR)
        print(f"FAISS index rebuilt with {total_chunks} chunks.")

        return IngestResponse(
            message=f"Successfully processed {pdf_file.filename} and rebuilt FAISS index.",
            total_chunks_processed=total_chunks,
            pdf_filename=pdf_file.filename
        )
    except Exception as e:
        import traceback
        traceback.print_exc()
        raise HTTPException(status_code=status.HTTP_500_INTERNAL_SERVER_ERROR, detail=f"Failed to ingest document: {e}")
    finally:
        if os.path.exists(temp_pdf_path):
            os.remove(temp_pdf_path)
            print(f"Cleaned up temporary PDF file: {temp_pdf_path}")

@app.post("/query", response_model=QueryResult, status_code=status.HTTP_200_OK, summary="Query RAG System")
async def query_rag(request: QueryRequest):
    """Queries the RAG system to get an answer based on the ingested document and cited sources."""
    from src.retrieval.vector_db import faiss_index, all_chunks, embedding_model
    from src.retrieval.rag_pipeline import OPENAI_API_KEY

    if faiss_index is None or not all_chunks or embedding_model is None:
        raise HTTPException(
            status_code=status.HTTP_503_SERVICE_UNAVAILABLE,
            detail="RAG index not initialized. Please ingest a document first via the /ingest endpoint."
        )
    if OPENAI_API_KEY is None:
        raise HTTPException(
            status_code=status.HTTP_503_SERVICE_UNAVAILABLE,
            detail="OpenAI API key is not configured. Please set it in Colab secrets."
        )

    try:
        answer, sources_list = answer_query_with_rag_fastapi(request.query_string, request.k)
        if answer.startswith("Error communicating with LLM"):
             raise HTTPException(status_code=status.HTTP_500_INTERNAL_SERVER_ERROR, detail=answer)
        return QueryResult(answer=answer, sources=sources_list)
    except RuntimeError as e:
        raise HTTPException(status_code=status.HTTP_500_INTERNAL_SERVER_ERROR, detail=f"RAG system error: {e}")
    except Exception as e:
        import traceback
        traceback.print_exc()
        raise HTTPException(status_code=status.HTTP_500_INTERNAL_SERVER_ERROR, detail=f"An unexpected error occurred during query: {e}")

# To run the FastAPI application in Colab, you would typically use ngrok or similar.
# For local testing within Colab, you can run:
# if __name__ == "__main__":
#     import uvicorn
#     uvicorn.run(app, host="0.0.0.0", port=8000)
````
